### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui as pg
# pg.click -> clicar
# pg.write -> digitar   
# pg.press -> apertar tecla
#pg.hotkey -> apertar varias teclas juntas
#pg.sleep -> esperar

pg.press("win")
pg.write("mozilla")
pg.sleep(1)
pg.press("enter")
pg.sleep(3)
pg.write("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga")
pg.press("enter")
pg.sleep(3)
pg.click(x=466, y=528 , clicks = 2)  # clicar em exportar
pg.sleep(3)
pg.click(x=1715, y=531)  # clicar para baixar


```

```python
import pandas as pd
caminho = r"C:\Users\diego\Downloads\Vendas - Dez.xlsx" 
tabela = pd.read_excel(caminho)


#somar as vendas
total_vendas = tabela["Valor Final"].sum()
#quantidade de produtos vendidos
quantidade_produtos = tabela["Quantidade"].sum()
print(total_vendas)
print(quantidade_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyautogui as pg
import time
import pyperclip  
import locale

# 1. Configurar Locale (Brasil)
try:
    locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
except:
    locale.setlocale(locale.LC_ALL, 'pt_BR')

# -----------------------------------------------------------------

# 2. Preparar o texto ANTES de começar a automação
# Aqui formatamos o dinheiro e a quantidade corretamente


texto_email = f"""Prezados, bom dia!

Segue o relatório de vendas de dezembro.

O total de vendas foi de {total_vendas:,.2f} reais.
A quantidade de produtos vendidos foi de {quantidade_produtos:,} unidades.

Qualquer dúvida estou à disposição.

Att., Diego Menezes"""

# --- INÍCIO DA AUTOMAÇÃO ---

pg.press("win")
pg.write("mozilla") 
pg.sleep(1)
pg.press("enter")
pg.sleep(3)

# Enviar email
pg.hotkey("ctrl", "t")
pg.sleep(1)
pg.write("https://mail.google.com/mail/u/0/#inbox")
pg.sleep(1)
pg.press("enter")   
pg.sleep(5) 

pg.click(x=165, y=215)  # Clicar em novo email
pg.sleep(3) 

# Destinatário
pg.write("diegomenez@gmail.com" , interval=0.2) 
pg.sleep(1)
pg.press("tab", presses=2) # Pular para o assunto    
pg.sleep(1)

# Assunto (Como é curto e sem acentos complexos, o write funciona bem, mas o 'ó' pode falhar)
# Se o 'ó' falhar aqui, use pyperclip aqui também.
pg.write("Relatorio de Vendas de Dezembro") 
pg.press("tab") # Pular para o corpo do email     
pg.sleep(1)

# --- CORREÇÃO PRINCIPAL AQUI ---
# Em vez de escrever letra por letra (que falha nos acentos), colamos o texto pronto
pyperclip.copy(texto_email)
pg.hotkey("ctrl", "v")
pg.sleep(1)
pg.click(x=1459, y=1002)  # Clicar em anexo
pg.sleep(2)
pg.write(caminho)
pg.sleep(1)
pg.press("enter")
pg.sleep(3) 
pg.hotkey("ctrl", "enter")  # Enviar o email
```
